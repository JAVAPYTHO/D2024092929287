```python
import pandas as pd

def load_data(file):
    try:
        # 尝试以 UTF-8 读取文件
        df = pd.read_csv(file, encoding='utf-8')
    except UnicodeDecodeError:
        try:
            # 尝试以 ISO-8859-1 读取文件
            df = pd.read_csv(file, encoding='ISO-8859-1')
        except UnicodeDecodeError:
            # 尝试以 GBK 读取文件（常见于中文文件）
            df = pd.read_csv(file, encoding='GBK')
    
    # 清理列名：去掉回车符和空格
    df.columns = df.columns.str.replace(r'\r', '', regex=True).str.strip()
    
    # 将“小类”列名替换为“type”
    df.rename(columns={'小类': 'type'}, inplace=True)
    
    # 去除重复值
    df.drop_duplicates(inplace=True)
    
    # 检查并处理缺失值
    df.fillna(method='ffill', inplace=True)
    
    return df


import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_transportation_facilities(file):
    # 加载数据
    df = load_data(file)
    
    # 可视化交通设施的经纬度分布
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x='lon_gcj02', y='lat_gcj02', hue='type', palette='Set1', s=50, edgecolor='w', alpha=0.7, legend=False)
    
    # 设置标题和标签
    plt.title("Distribution of Transportation Facilities", fontsize=16)
    plt.xlabel("Longitude (lon_gcj02)", fontsize=12)
    plt.ylabel("Latitude (lat_gcj02)", fontsize=12)
    
    # 显示图例
    # plt.legend(title='Facility Type', bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # 显示图形
    plt.tight_layout()
    plt.show()

# 示例：使用交通设施数据文件
plot_transportation_facilities('F:\pythonProject\python\question2/Appendix 3/Transportation facilities data.csv')
```
